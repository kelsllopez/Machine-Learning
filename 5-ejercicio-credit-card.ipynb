```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier  # O el clasificador que estés usando
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score

# Cargar y preparar tus datos
import pandas as pd
df = pd.read_csv('/kaggle/input/creditcardfraud/creditcard.csv')

# Separar características y etiqueta
X = df.drop('Class', axis=1)
y = df['Class']

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el modelo
model = RandomForestClassifier(random_state=42)  # Usa el clasificador que prefieras
model.fit(X_train, y_train)

# Realizar predicciones
y_pred = model.predict(X_test)

# Calcular métricas
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

# Mostrar resultados
print(f'Precisión: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'Puntaje F1: {f1:.2f}')
print(f'Exactitud: {accuracy:.2f}')
```
