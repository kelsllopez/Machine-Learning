```python
# Obtenemos el dataset que usaremos.
df = pd.read_csv('/kaggle/input/dataset-houses/housing.csv')

# Identificamos valores nulos.
print(df.isnull().sum())


```

```python
# Eliminar filas con valores nulos en todo el DataFrame
df_clean = df.dropna()
# Verificar valores nulos después de la eliminación
print(df_clean.isnull().sum())
```

```python
# Vamos a hacer un analisis de regresion lineal multiple
from sklearn.linear_model import LinearRegression
import pandas as pd

# Preparamos los datos una vez ya limpios.
# X contiene las características de las casas.
# Y contiene el valor de las casas que queremos predecir.
X = df_clean[['housing_median_age', 'total_rooms', 'total_bedrooms', 'population', 'households', 'median_income']]
y = df_clean['median_house_value']



# Ajustar el modelo 
# se crea un modelo de regresión lineal y se ajusta con X (características) y y (valores).
model = LinearRegression()
model.fit(X, y)



# Obtener coeficientes
# coefficients son los valores que indican el impacto de cada característica en el valor de la casa.
coefficients = model.coef_
features = X.columns

# Mostrar resultados
# coef_df organiza estos coeficientes en un DataFrame para facilitar su interpretación
coef_df = pd.DataFrame(coefficients, index=features, columns=['Coefficient'])
coef_df = coef_df.sort_values(by='Coefficient', ascending=False)
print(coef_df)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Hacer predicciones
y_pred = model.predict(X)

# Calcular el R² (coeficiente de determinación)
r2 = r2_score(y, y_pred)

# Calcular el Error Cuadrático Medio (MSE)
mse = mean_squared_error(y, y_pred)

# Calcular la Raíz del Error Cuadrático Medio (RMSE)
rmse = np.sqrt(mse)

# Calcular el Error Absoluto Medio (MAE)
mae = mean_absolute_error(y, y_pred)

# Mostrar resultados
print(f'MAE (Error Absoluto Medio): {mae}')
print(f'MSE (Error Cuadrático Medio): {mse}')
print(f'R² (R al cuadrado): {r2}')
print(f'RMSE (Raíz del Error Cuadrático Medio): {rmse}')

```
