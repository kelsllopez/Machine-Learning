```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Cargar los datos de las películas y calificaciones
movie = pd.read_csv('/kaggle/input/peliculas/movie.csv')
rating = pd.read_csv('/kaggle/input/peliculas/rating.csv')
```

# Filtrar las películas de terror


```python
peliculas_de_terror = movie[movie['genres'].str.contains('Horror', case=False, na=False)]

# Ver las primeras películas de terror
peliculas_de_terror.head()
```

# Mostar las mejores peliculas de terror con su rating

```python
peliculas_de_terror = pd.merge(rating, peliculas_de_terror, on='movieId')

# Calcular el promedio de calificaciones para cada película de terror
peliculas_de_terror_con_su_rating = peliculas_de_terror.groupby(['title', 'genres'])['rating'].mean().sort_values(ascending=False)

# Mostrar las 5 mejores películas de terror con las calificaciones más altas
top_5_peliculas_de_terror = peliculas_de_terror_con_su_rating.head(5)

# Imprimir los resultados
print(top_5_peliculas_de_terror)
```

# Filtrar para obtener el género de Toy Story directamente


```python
genero_toy_story = movie[movie['title'].str.contains('Toy Story', case=False, na=False)]['genres'].values[0]

print(f"El género de la película Toy Story es: {genero_toy_story}")
```

# filtar pelicula por el genero de la pelicula toy story , 
# en la cual la clasificaremos por el rating y el genero de ella y poniendo de mayor a menor

```python
peliculas_similares_por_genero = movie[movie['genres'].str.contains('Adventure|Animation|Children|Comedy|Fantasy', case=False, na=False)]

peliculas_similares_por_genero_y_rating = pd.merge(peliculas_similares_por_genero, rating, on='movieId')

peliculas_similares_por_rating = peliculas_similares_por_genero_y_rating.groupby(['title', 'genres'])['rating'].mean().sort_values(ascending=False)

peliculas_similares_por_rating.head()
```
