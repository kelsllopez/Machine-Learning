# Dividir el DataFrame en Características y Etiquetas

```python
# Dividir en características y etiquetas
X = df.iloc[:, :-1]  # Características (todas las columnas excepto la última)
y = df.iloc[:, -1]   # Etiqueta (columna 57, spam o no spam)
```

# Dividir los Datos en Conjuntos de Entrenamiento y Prueba

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Escalar las Características

#  ** StandardScaler  puede ayudar a que el árbol de decisión funcione de manera más eficiente.**

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Entrenar el Modelo

```python
dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train_scaled, y_train)
```

# Realizar Predicciones

```python
y_pred_dtc = dtc.predict(X_test_scaled)
```

# Evaluar el Modelo

```python
# Calcular métricas
Exactitud = accuracy_score(y_test, y_pred_dtc)
tasa_de_error = 1 - accuracy
puntaje_f1 = f1_score(y_test, y_pred_dtc)
puntaje_precision = precision_score(y_test, y_pred_dtc)
matrix_confusion = confusion_matrix(y_test, y_pred_dtc)

# Imprimir resultados
# Imprimir resultados
print("Reporte de Clasificación del Árbol de Decisión\n")
print(classification_report(y_test, y_pred_dtc))
print("Exactitud(Accuracy):", Exactitud)
print("Tasa de Error(Error Rate):", tasa_de_error)
print("Puntaje F1(F1-Score):", puntaje_f1)
print("Puntaje de Precisión(Precision Score):", puntaje_precision)
print("Matriz de Confusión(Confusion Matrix):")
print(matrix_confusion)
```
